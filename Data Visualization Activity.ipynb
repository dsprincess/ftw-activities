## Import Datasets

We are going to us the [Iris](https://archive.ics.uci.edu/ml/datasets/iris) and the [Wine Reviews](https://www.kaggle.com/zynicide/wine-reviews) dataset.

```python
%matplotlib inline  
import pandas as pd
import numpy as np
```

```python
# import Data_Extract_From_World_Development_Indicators data frame with the .. values as null
# filter dataframe by Series Code of SH.STA.BASS.ZS: People using at least basic sanitation services (% of population)
worlddev = pd.read_excel('Data_Extract_From_World_Development_Indicators.xlsx', sheet_name = 'Data', na_values='..')
```

```python
worlddev.info()
worlddev.describe()
```

Subsets

```python
popsani=worlddev.loc[worlddev['Series Code'] == 'SH.STA.BASS.ZS']
#return popsani dataframe
popsani
```

```python
popsani2 = worlddev.loc[worlddev['Series Code'] == 'SH.STA.BASS.ZS'].drop(['Country Code', 'Series Name', 'Series Code', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]'], axis=1).set_index('Country Name').fillna(0).T
popsani2
# .T or .transpose() to transpose the years as index
# dropped 2018-2020 since there's no data
```

```python
#dataset for Basic Sanitation % for Rural and Urban
options=['SH.STA.BASS.RU.ZS', 'SH.STA.BASS.UR.ZS']
popsaniRU=worlddev.loc[worlddev['Series Code'].isin(options)] 
popsaniRU
```

```python
#dataset for Consolidated yearly percentage of Basic Sanitation for each country
popsani2010 = popsani[["Country Name", "2010 [YR2010]"]]
popsani2011 = popsani[["Country Name", "2011 [YR2011]"]]
popsani2012 = popsani[["Country Name", "2012 [YR2012]"]]
popsani2013 = popsani[["Country Name", "2013 [YR2013]"]]
popsani2014 = popsani[["Country Name", "2014 [YR2014]"]]
popsani2015 = popsani[["Country Name", "2015 [YR2015]"]]
popsani2016 = popsani[["Country Name", "2016 [YR2016]"]]
popsani2017 = popsani[["Country Name", "2017 [YR2017]"]]
popsaniperyear= pd.DataFrame(np.concatenate([popsani2010.values, popsani2011.values,  popsani2012.values,  popsani2013.values,  popsani2014.values,  popsani2015.values,  popsani2016.values,  popsani2017.values], axis=0), columns=popsani2010.columns)
popsaniperyear.rename(columns={'2010 [YR2010]': 'Percentage'}, inplace=True)
popsaniperyear
```

## Matplotlib

Matplotlib is a Python plotting libary.

```python
import matplotlib.pyplot as plt
```

### Scatter plot

```python
# create a figure and axis
fig, ax = plt.subplots()
plt.rcdefaults()
Colorvariable = ['r','g','b','c','y','m','k',	"#4DBEEE", "#D95319", "#7E2F8E"]

# scatter the population % in 2010 versus 2012
ax.scatter(popsani['2010 [YR2010]'], popsani['2015 [YR2015]'],c=Colorvariable)
# set a title and labels
ax.set_title('2010 VS 2015')
ax.set_xlabel('2010')
ax.set_ylabel('2015')

# the 10 dots represent the 10 countries 
# comparison of each of the countries basic sanitation percentage 2010 vs 2015
# as seen on the figure, Myanmar on yellow dot, have a 2015 percentage that is lower than its 2010 unlike with all the other countries that increased. 
```

We can color each point by class

#### *Tweak*

```python
from matplotlib import rcParams, cycler
```

### Line Chart

```python
columns = popsani.columns.drop(['Country Name', 'Country Code', 'Series Name', 'Series Code'])

rcParams['lines.linewidth'] = 4
rcParams['axes.prop_cycle'] = cycler(color=['gray'])

fig, ax = plt.subplots()
for column in columns:
    ax.plot(popsani[column])
ax.set_title('People Using Basic Sanitation Services in %')
#ax.legend(popsani[columns])
popsani['2010 [YR2010]'].plot(color=['red'])
popsani['2015 [YR2015]'].plot(color=['blue'])

# There are 10 lines representing (2010 to 2017) 
# Cambodia on the Cambodia(2nd), Lao(3rd) and Indonesia(10th) were seen to have progress in its yearly percentage
```

```python
columns = popsani2

fig, ax = plt.subplots()

for column in columns:
    ax.plot(popsani2.index, popsani2[column])
    
ax.set_title('People Using Basic Sanitation Services 2010-2017')
ax.legend(popsani2.columns.values, bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.xticks(rotation = 60)

# Brunei has no data in 2016 and 2017 so it dropped
```

## Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

### Scatter plot

```python
sns.scatterplot(x='Country Name', y='Percentage', data=popsaniperyear)
plt.xticks(rotation=90)
```

Highlighting the classes is significally easier than using Matplotlib. We only need to specify the hue parameter.

```python
rcParams['figure.figsize'] = 8,7
sns.scatterplot(x='Country Name', y='Percentage', hue='Country Name', data=popsaniperyear)
plt.xticks(rotation=90)

#Cambodia have mostly scattered/varying percentages 2010 to 2015
#Brunei Darussalam and Singapore have consistent percentage
```

### Heatmap

Heatmaps are perfect for exploring the correlation of the features

```python
corr = popsani.corr()
fig, ax = plt.subplots()
im = ax.imshow(corr.values)
```

```python
#FacetGrid

g = sns.FacetGrid(popsaniRU, col='Series Name')
g = g.map(sns.kdeplot, '2010 [YR2010]')

axes = g.axes.flatten()
axes[0].set_title("% of Basic Sanitation-Rural")
axes[1].set_title("% of Basic Sanitation-Urban")

#For year 2010, Rural areas have wide varying range of percentages while Urban have closer percentages
```

```python
sns.pairplot(popsani)
```

```python
#Violin plot
figure, ax = plt.subplots()
sns.set(style = 'whitegrid')
    
sns.violinplot(x ="Series Name",
             y ="2017 [YR2017]",
             data = popsaniRU)

ax.set_xticklabels( ('Rural', 'Urban') )

#For year 2017, Rural areas have wide varying range of percentages while Urban have closer percentages
```

```python
#swarmplot
    
sns.swarmplot(y ="Country Name",
             x ="2010 [YR2010]",
             hue ="Series Name",
             data = popsaniRU)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#It can be seen here that Malaysia and Thailand's percentage of basic sanitation service for their rural and urban areas have closer percentage
```

```python
#factor plot (old name of catplot)
sns.factorplot(x ="Country Name",
             y ="2010 [YR2010]", 
               hue = 'Series Name', 
               data = popsaniRU)
plt.xticks(rotation=90)
```

```python
#stripplot 
#almost the same with swarmplot, the difference is that the points do not overlap  in swarmplot while in strip it overlaps
#because our df have small dataset, distinction between these two plots can't be demonstrated)
   
sns.swarmplot(y ="Country Name",
             x ="2010 [YR2010]",
             hue ="Series Name",
             data = popsaniRU)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

#It can be seen here that Malaysia and Thailand's basic sanitation services for their rural and urban areas have closer percentage
```
