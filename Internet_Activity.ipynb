```python
world_df = pd.read_csv('https://raw.githubusercontent.com/angelaaaateng/ftw_python/main/data/Data_Extract_From_World_Development_Indicators.csv')
world_df
```

```python
internet_df = world_df[world_df['Series Code'] == 'IT.NET.USER.ZS']
internet_df = internet_df.replace("..", "0")
internet_df[cols_to_avg] = internet_df[cols_to_avg].apply(pd.to_numeric, errors='coerce')
internet_df = internet_df.set_index('Country Name')
internet_df = internet_df[cols_to_avg]
internet_df
```

```python
internet_dfnonzero = internet_df.replace(0, np.NaN)
internet_dfnonzero.mean()
```

```python
#Describe the YoY (Year on Year) Growth (or decline) of internet users (%) overall.
internet_dfO = pd.DataFrame(internet_dfnonzero.mean())
internet_dfO
```

```python
internet_dfO.rename(columns = {0:'Average Percentage of Internet Users'}, inplace = True)
internet_dfO.index.names = ['Year']
internet_dfO.plot(figsize=(12, 8), title='Average Percentage of Internet Users Over Time (Total)')

plt.savefig('Average Percentage of Internet Users Overall.jpg', bbox_inches='tight')
```

```python
internetyoyC_df = pd.DataFrame(internet_dfnonzero.mean(axis=1))
internetyoyC_df
```

```python
internetyoyC_df.rename(columns = {0:'Average Percentage of Internet Users Per Country'}, inplace = True)
sorted_df = internetyoyC_df.sort_values(by='Average Percentage of Internet Users Per Country', ascending=False)
sorted_df
```

```python
sorted_df.plot.bar()
plt.savefig('Internet Users Per Country.jpg', bbox_inches='tight')
```

```python
internet_dfPCPY = internet_df.T
internet_dfPCPY = internet_dfPCPY.reset_index()
internet_dfPCPY
```
